```python
# Pie Chart for Shipping Method
fig = px.pie(df_uq, names='Shipping_Method', title='Pie Chart of Shipping Method')

# Update layout with black background and white font
fig.update_layout(
    paper_bgcolor="black", 
    font_color="white", 
    legend_font_color="white", 
    title_font_color="white"
)

# Display the plot
fig.show()
```

```python
# Pie Chart for Category
fig = px.pie(df_uq, names='Category', title='Pie Chart of Category')

# Update layout with black background and white font
fig.update_layout(
    paper_bgcolor="black", 
    font_color="white", 
    legend_font_color="white", 
    title_font_color="white"
)

# Display the plot
fig.show()
```

```python
# Bar graph for Payment_Terms

# Default selection for Payment_Terms
default_selection_payment_terms = ['Payment_Terms']

# Dropdown for selecting categorical variable (Payment Terms)
payment_terms_dropdown = widgets.Dropdown(
    options=default_selection_payment_terms,
    value=default_selection_payment_terms[0],
    description='Payment Terms:',
    disabled=False
)

# Function to update and display the bar chart for Payment Terms
def update_bar_chart_payment_terms(selected_var):
    # Ensure that the selected variable exists in df_uq
    if selected_var in df_uq.columns:
        # Count occurrences of each category in Payment_Terms
        category_counts = df_uq[selected_var].value_counts().reset_index()
        category_counts.columns = [selected_var, 'Count']  # Renaming columns for better understanding
        
        # Create bar chart with different colors for each category
        fig = px.bar(
            category_counts, 
            x=selected_var, 
            y='Count', 
            color=selected_var,  # Assign different colors for categories
            title=f"Bar Chart - {selected_var}",
            color_discrete_sequence=px.colors.qualitative.Bold  # Set color scheme
        )
        fig.update_layout(bargap=0.5)  # Adjust the space between bars to reduce bar width
        fig.show()
    else:
        print(f"{selected_var} not found in the dataset")

# Create interactive output for bar chart (Payment Terms)
output_bar_payment_terms = widgets.interactive_output(update_bar_chart_payment_terms, {'selected_var': payment_terms_dropdown})

# Display dropdown and bar chart for Payment Terms
display(payment_terms_dropdown, output_bar_payment_terms)
```

```python
# Bar graph for Import_Export

# Default selection for Import_Export
default_selection_import_export = ['Import_Export']

# Dropdown for selecting categorical variable (Import/Export)
import_export_dropdown = widgets.Dropdown(
    options=default_selection_import_export,
    value=default_selection_import_export[0],
    description='Import/Export:',
    disabled=False
)

# Function to update and display the bar chart for Import/Export
def update_bar_chart_import_export(selected_var):
    # Ensure that the selected variable exists in df_uq
    if selected_var in df_uq.columns:
        # Count occurrences of each category in Import_Export
        category_counts = df_uq[selected_var].value_counts().reset_index()
        category_counts.columns = [selected_var, 'Count']  # Renaming columns for better understanding
        
        # Create bar chart with different colors for each category
        fig = px.bar(
            category_counts, 
            x=selected_var, 
            y='Count', 
            color=selected_var,  # Assign different colors for categories
            title=f"Bar Chart - {selected_var}",
            color_discrete_sequence=px.colors.qualitative.Vivid  # Set color scheme
        )
        fig.update_layout(bargap=0.7)  # Adjust the space between bars to reduce bar width
        fig.show()
    else:
        print(f"{selected_var} not found in the dataset")

# Create interactive output for bar chart (Import/Export)
output_bar_import_export = widgets.interactive_output(update_bar_chart_import_export, {'selected_var': import_export_dropdown})

# Display dropdown and bar chart for Import/Export
display(import_export_dropdown, output_bar_import_export)
```

```python
# Line chart of Import_Export over time

# Convert the 'Date' column to datetime format, specify the format
df_uq['Date'] = pd.to_datetime(df_uq['Date'], format='%m/%d/%Y', errors='coerce')

# Check for any conversion issues
if df_uq['Date'].isnull().any():
    print("There are some invalid dates in the DataFrame.")

# Extract the year from the 'Date' column
df_uq['Year'] = df_uq['Date'].dt.year

# Aggregate data by year and Import/Export category
import_export_summary = df_uq.groupby(['Year', 'Import_Export']).agg({'Quantity': 'sum'}).reset_index()

# Create line chart using Plotly with vibrant colors
fig = px.line(import_export_summary, x='Year', y='Quantity', color='Import_Export', 
              title='Line Chart of Imports and Exports Over Time', markers=True,
              color_discrete_map={'Import': '#FF4500', 'Export': '#1E90FF'})  # Tomato and Lime Green

# Update layout with black background and white font
fig.update_layout(
    paper_bgcolor="black", 
    plot_bgcolor="white",  # Set plot background to black
    font_color="white", 
    legend_font_color="white", 
    title_font_color="white",
    xaxis=dict(gridcolor='black'),  # Set x-axis gridlines to black
    yaxis=dict(gridcolor='black')   # Set y-axis gridlines to black
)

# Display the plot
fig.show()
```

```python
# Scatter Plot of Quantity vs Value (Imports and Exports)

# Create a scatter plot using Plotly
fig = px.scatter(df_uq, x='Quantity', y='Value', color='Import_Export', 
                 title='Scatter Plot of Quantity vs Value (Imports and Exports)',
                 color_discrete_map={'Import': 'darkblue', 'Export': 'red'}) 

# Update layout with black background, white font, black gridlines
fig.update_layout(
    paper_bgcolor="black", 
    plot_bgcolor="black",  # Set plot background to black
    font_color="white", 
    legend_font_color="white", 
    title_font_color="white",
    xaxis=dict(gridcolor='white'),
    yaxis=dict(gridcolor='white'),   
)

# Display the plot
fig.show()
```

```python
# Box plot of Shipping Method

# Create a box plot using Plotly
fig = px.box(df_uq, x='Shipping_Method', y='Value', title="Box Plot of Value Distribution by Shipping Method",
             color='Shipping_Method',  # Use Shipping_Method for coloring
             color_discrete_map={
                 'Air': '#ADD8E6',   # Light Blue
                 'Sea': '#90EE90',   # Light Green
                 'Land': '#FFD700'   # Yellow
             })

# Update layout with black background, white font, and gridlines
fig.update_layout(
    paper_bgcolor="black",
    plot_bgcolor="black",  # Set plot background to black
    font_color="white",
    legend_font_color="white",
    title_font_color="white",
    xaxis=dict(gridcolor='white', zerolinecolor='black'),  # Set x-axis gridlines to white, zero line to black
    yaxis=dict(gridcolor='white', zerolinecolor='black')   # Set y-axis gridlines to white, zero line to black
)

# Display the plot
fig.show()
```

```python
# Heatmap

# Assuming numeric_options includes all columns
numeric_options = df_uq.columns.tolist()

# Select numeric variables for heatmap (replace with your user-selected values)
selected_heatmap_vars = numeric_options  # For demonstration

# Filter selected variables to include only numeric types
selected_heatmap_vars = [col for col in selected_heatmap_vars if df_uq[col].dtype in [np.float64, np.int64]]

# Convert non-numeric columns to numeric, coercing errors to NaN
for col in selected_heatmap_vars:
    df_uq[col] = pd.to_numeric(df_uq[col], errors='coerce')

# Drop NaN values if any remain
df_uq.dropna(subset=selected_heatmap_vars, inplace=True)

# Calculate correlation matrix
correlation_matrix = df_uq[selected_heatmap_vars].corr()

# Create heatmap using Plotly
fig = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.index,
    colorscale='Viridis',  # Vibrant color scale
    colorbar=dict(title="Correlation"),
    hovertemplate='Variable 1: %{x}<br>Variable 2: %{y}<br>Correlation: %{z:.2f}<extra></extra>'
))

# Update layout
fig.update_layout(
    title="Heatmap of Selected Variables",
    paper_bgcolor="black",
    plot_bgcolor="black",
    font_color="white",
    title_font_color="white",
    xaxis=dict(titlefont=dict(color='white'), tickfont=dict(color='white')),
    yaxis=dict(titlefont=dict(color='white'), tickfont=dict(color='white')),
)

# Display the heatmap
fig.show()
```

```python
# Histogram of Quantity, Value and Weight

# Define numeric options based on your DataFrame columns
numeric_options = df_uq.select_dtypes(include=['number']).columns.tolist()  # Select only numeric columns

# Create histograms for Value, Quantity, and Weight with different colors
hist_colors = {
    'Value': 'blue',
    'Quantity': 'orange',
    'Weight': 'green'
}

# Loop through each numeric variable to create histograms
for hist_var in numeric_options:
    if hist_var in hist_colors:  # Check if the variable has a defined color
        fig = px.histogram(
            df_uq, 
            x=hist_var, 
            nbins=30, 
            title=f'Histogram of {hist_var}', 
            marginal="box",
            color_discrete_sequence=[hist_colors[hist_var]]  # Set color for each variable
        )
        
        # Update layout for better visibility
        fig.update_layout(
            paper_bgcolor="black", 
            plot_bgcolor="black",
            font_color="white", 
            legend_font_color="white", 
            title_font_color="white"
        )
        
        # Show the histogram
        fig.show()
```
